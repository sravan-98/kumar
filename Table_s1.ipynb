```python
import pandas as pd
import seaborn as sns
```

```python
dt=pd.read_csv('//content/archive (1).zip')
```

```python
dt.info()
```

```python
for i in dt.columns:
  if dt[i].dtype == 'object':
    dt[i]=dt[i].astype('category').cat.codes
```

```python
dt.info()
```

```python
dt.drop(['customerID'],axis = 1, inplace=True)
```

```python
dt.corr()
```

```python
dt = dt.drop(["gender","Dependents","PhoneService","MultipleLines","InternetService"],axis=1)
```

```python
dt=dt.drop(["StreamingTV","StreamingMovies","TotalCharges"],axis=1)
```

```python
x=dt.drop("Churn",axis=1)
y=dt["Churn"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
sns.pairplot(data = dt,hue = 'Churn')
```

```python
from sklearn.tree import DecisionTreeClassifier
d = DecisionTreeClassifier()
```

```python
d.fit(xtrain,ytrain)
```

```python
ypred=d.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```
