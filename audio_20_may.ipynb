```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
ip = "/content/image.png"
plt.imshow(Image.open(ip))
plt.axis("off")
```

```python
img = Image.open(ip)
text = pytesseract.image_to_string(img)
print("Extracted text:",text)
```

```python
from gtts import gTTS
b=gTTS(text)
b.save("/content/image.png")
```
